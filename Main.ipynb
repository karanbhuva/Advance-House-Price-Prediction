# Create data for train and test

```python
x_train=train.iloc[:,1:train.shape[1]-1]
y_train=train.iloc[:,train.shape[1]-1]
x_test=test.iloc[:,1:test.shape[1]]
```

# check whether train and test features are same or different

```python
c=0
for i in range(x_train.shape[1]):
    if x_train.columns[i]==x_test.columns[i]:
        c=c+1
print(c,len(x_test.columns))
```

# Apply Regression on data

```python

 
# create regressor object 
regressor = RandomForestRegressor(n_estimators = 500, random_state = 0) 
 
# fit the regressor with x and y data 
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
```

```python
prediction=regressor.predict(x_test)
```
